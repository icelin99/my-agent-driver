```python
from nuscenes.nuscenes import NuScenes
nusc = NuScenes(version='v1.0-trainval', dataroot='./DataSets/nuScenes', verbose=True)
```

```python
index_list = []  
with open('./sample-index.txt', 'r') as file:
    for line in file:
        index = int(line.strip())  # 去除可能的空白字符，并转换为整数
        index_list.append(index)
```

```python
sample = nusc.sample[index_list[100]]
sam_front_token = sample['data']['CAM_FRONT']
sam_ra_front_token = sample['data']['LIDAR_TOP']
```

```python
nusc.render_sample_data(sam_ra_front_token)
```
